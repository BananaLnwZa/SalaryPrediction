```python
import numpy as np
import pandas as pd
```

```python
# Data Loading and Exploration
data_path = 'C:/Users/User/Documents/Salary_Prediction/backend/data/Salary Data.csv'

# Reading the CSV file with the specified encoding
df = pd.read_csv(data_path, encoding='UTF-8-SIG')

# Display the first few rows of the dataset
print('First few rows of the dataset:')
display(df.head())

# Display summary of the dataset
print('\nDataset info:')
df.info()

print('\nStatistical summary:')
display(df.describe(include='all'))
```

```python
# Data Cleaning and Preprocessing
# It appears that missing values in the features are causing errors in model fitting.
# In this notebook, we resolve this by dropping rows with missing values.

print('Missing values per column before cleaning:')
display(df.isnull().sum())

# If you prefer to impute missing values, you can use sklearn's SimpleImputer or fillna methods.
df.dropna(inplace=True)

print('Missing values per column after cleaning:')
display(df.isnull().sum())

# Checking data types
print('Data types:')
display(df.dtypes)

# If there were any date columns, we would convert them to datetime. (None in this dataset)

df.to_csv("NewSalaryData.csv", index=False)
```
